Constructing a Quantum Circuit

1. applying gates and visualizing the circuit.

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt


qc = QuantumCircuit(5)

qc.h(0)      # HADAMARD GATE
qc.x(1)      # PAULI -X GATE
qc.z(2)      # PAULI -Z GATE
qc.y(3)      # PAULI -Y GATE
qc.cx(1,0)   # CNOT GATE / CONTROLLED NOT / CONTROLLED X
qc.s(4)      # S GATE (PHASE GATE)
qc.ccx(1,2,3)  # TAFFOLI GATE
qc.swap(2,3)  # SWAP GATE
qc.measure_all()   # i(|01010⟩ + |11110⟩)/√2 after applying all gates this will bee the final state measurment
qc.draw(output="mpl")
```

2. Simulating the circuit and printing the measurement

```python

from qiskit import transpile
from qiskit_aer import AerSimulator
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
result=simulator.run(compiled_circuit, shots=4096).result()
counts = result.get_counts()
print("Measurement Results:", counts)
```

this shows that both 00111 and 001100 have  equal probabilities of occurrence

### ploting histogram

```python
plot_histogram(counts)
```
